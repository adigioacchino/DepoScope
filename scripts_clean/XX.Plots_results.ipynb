```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os 

path_figures = "/media/concha-eloko/Linux/PhageDEPOdetection/figures"
folds_results = [("beta-helix", 828) , ("Time beta/alpha barrel", 76) , ("n-bladed beta propeller", 150), ("alpha/alpha toroid",40), ("Other",6)]

group_names, numbers = zip(*folds_results)

group_names , numbers = list(group_names) , list(numbers)
# Set up the Seaborn style
sns.set(style="whitegrid")

# Create a bar plot using Seaborn
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
sns.barplot(x=numbers, y=group_names)

plt.xlabel('Number')
plt.ylabel('Depolymerase fold')
plt.title('Bar Plot of Group Numbers')
plt.savefig(f"{path_figures}/Figure_count_folds.svg", dpi = 600)
plt.show()

```

```python
plt.figure(figsize=(8, 8))  # Adjust the figure size as needed

plt.pie(numbers, labels=group_names, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')
# Add a title
plt.title('Pie Chart of Group Numbers')

plt.show()
```
